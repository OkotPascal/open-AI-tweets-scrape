```python
scrapper = sntwitter.TwitterSearchScraper('#OpenAI')

tweets = []

for _, tweet in enumerate (scrapper.get_items()):
    
    data = [tweet.rawContent, tweet.renderedContent, tweet.user.displayname, 
            tweet.user.link, tweet.retweetedTweet, tweet.user.followersCount, 
            tweet.user.verified, tweet.user.created, tweet.user.location]
    
    tweets.append(data)
    
    if _ > 100:
        break
```

```python
len(tweets)
```

```python
tweets
```

```python
tweets_df = pd.DataFrame(tweets, columns = ['tweet.rawContent', 'tweet.renderedContent', 'tweet.user.displayname', 
            'tweet.user.link', 'tweet.retweetedTweet', 'tweet.user.followersCount', 
            'tweet.user.verified', 'tweet.user.created', 'tweet.user.location'])
tweets_df
```

```python
tweets_df.to_csv('Open_AI Tweets.csv', index=False)
```
